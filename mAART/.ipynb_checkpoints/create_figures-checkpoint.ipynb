```python
def auto_limits(x, y, padding_factor=0.25, default_lim=(-0.2, 0.2)):
    """
    Automatically calculate xlim and ylim values with divergence handling.
    
    Parameters:
    x (array-like): x-coordinate data
    y (array-like): y-coordinate data
    padding_factor (float): percentage of range to add as padding (default: 0.05)
    default_lim (tuple): fallback limits when data is divergent (default: (-0.2, 0.2))
    
    Returns:
    tuple: (xlim, ylim) where each is a (min, max) tuple
    """
    # Convert to numpy arrays
    x = np.asarray(x).real
    y = np.asarray(y).real
    
    def calculate_lim(data):
        # Handle empty data
        if len(data) == 0:
            return default_lim
        
        # Handle constant values
        if np.all(data == data[0]):
            center = data[0]
            return (center - 1, center + 1)  # Symmetric range around single value
        
        # Handle infinite or NaN values
        if not np.all(np.isfinite(data)):
            return default_lim
            
        # Calculate regular limits with padding
        data_min, data_max = np.nanmin(data), np.nanmax(data)
        data_range = data_max - data_min
        
        # Handle extremely large ranges (divergent cases)
        if data_range > 1e6:  # Adjust threshold as needed
            return default_lim
            
        padding = data_range * padding_factor
        return (data_min - padding, data_max + padding)
    
    xlim = calculate_lim(x)
    ylim = calculate_lim(y)
    
    return xlim, ylim
```

```python


def auto_limits(x1, y1, x2, y2, padding_factor=0.25, default_lim=(-0.2, 0.2)):
    """
    Automatically calculate xlim and ylim values for two sets of data with divergence handling.
    
    Parameters:
    x1 (array-like): x-coordinate data for first set
    y1 (array-like): y-coordinate data for first set
    x2 (array-like): x-coordinate data for second set
    y2 (array-like): y-coordinate data for second set
    padding_factor (float): percentage of range to add as padding (default: 0.25)
    default_lim (tuple): fallback limits when data is divergent (default: (-0.2, 0.2))
    
    Returns:
    tuple: (xlim, ylim) where each is a (min, max) tuple
    """
    # Convert to numpy arrays and take real parts
    x1 = np.asarray(x1).real
    y1 = np.asarray(y1).real
    x2 = np.asarray(x2).real
    y2 = np.asarray(y2).real
    
    def calculate_lim(data):
        # Handle empty data
        if len(data) == 0:
            return default_lim
        
        # Handle constant values
        if np.all(data == data[0]):
            center = data[0]
            return (center - 1, center + 1)  # Symmetric range around single value
        
        # Handle infinite or NaN values
        if not np.all(np.isfinite(data)):
            return default_lim
            
        # Calculate regular limits with padding
        data_min, data_max = np.nanmin(data), np.nanmax(data)
        data_range = data_max - data_min
        
        # Handle extremely large ranges (divergent cases)
        if data_range > 1e6:  # Adjust threshold as needed
            return default_lim
            
        padding = data_range * padding_factor
        return (data_min - padding, data_max + padding)
    
    # Combine x data from both sets for overall xlim
    xlim = calculate_lim(np.concatenate([x1, x2]))
    
    # Combine y data from both sets for overall ylim
    ylim = calculate_lim(np.concatenate([y1, y2]))
    
    return xlim, ylim
```

```python


def create_image_grid(images, row_titles, col_titles, filename='img.png', output_path='./', img_size=(400, 400), fig_size=(6, 6), resize_images=True, dpi=300, wspace=0.1, hspace=0.1):
    """
    Creates a 2x2 grid of images with row and column titles.
    - images: List of 4 image paths (strings) or PIL Image objects.
    - row_titles: List of 2 strings for row titles (displayed vertically on the left).
    - col_titles: List of 2 strings for column titles (displayed horizontally on top).
    - output_path: Path to save the output image (default: 'grid_output.png').
    - img_size: Tuple (width, height) to resize each image to if resize_images is True (default: (400, 400)).
    - fig_size: Tuple (width, height) for the figure size when resize_images is True (default: (10, 10)). Ignored when resize_images is False.
    - resize_images: Boolean to enable/disable resizing images to img_size (default: True). If False, images are displayed at their original sizes, and the grid adjusts to fit them.
    - dpi: Resolution for saving the image (default: 600). Higher values improve quality but increase file size.
    - wspace: Horizontal spacing between subplots (default: 0.1). Increase for more space, decrease for less.
    - hspace: Vertical spacing between subplots (default: 0.1). Increase for more space, decrease for less.
    """
    if len(images) != 4:
        raise ValueError("Exactly 4 images are required for a 2x2 grid.")
    if len(row_titles) != 2 or len(col_titles) != 2:
        raise ValueError("Exactly 2 row titles and 2 column titles are required.")
    
    # Load images and get their sizes
    imgs = []
    sizes = []
    for img in images:
        if isinstance(img, str):
            img = Image.open(img)
        sizes.append(img.size)  # (width, height)
        if resize_images:
            img = img.resize(img_size)
        imgs.append(np.array(img))
    
    if resize_images:
        # Use fixed fig_size and standard subplots with custom spacing
        fig, axes = plt.subplots(2, 2, figsize=fig_size, gridspec_kw={'wspace': wspace, 'hspace': hspace})
        for i in range(2):
            for j in range(2):
                axes[i, j].imshow(imgs[i * 2 + j])
                axes[i, j].axis('off')
        
        # Add titles as before
        for j in range(2):
            x_pos = 0.25 + j * 0.5
            fig.text(x_pos, 0.95, col_titles[j], ha='center', va='bottom', fontsize=12, fontweight='bold')
        for i in range(2):
            y_pos = 0.75 - i * 0.5
            fig.text(0.05, y_pos, row_titles[i], ha='center', va='center', rotation='vertical', fontsize=12, fontweight='bold')
        #plt.tight_layout(rect=[0.1, 0.1, 0.9, 0.9])
    else:
        # Calculate grid dimensions based on image sizes
        col_widths = [max(sizes[0][0], sizes[2][0]), max(sizes[1][0], sizes[3][0])]
        row_heights = [max(sizes[0][1], sizes[1][1]), max(sizes[2][1], sizes[3][1])]
        
        total_width = sum(col_widths)
        total_height = sum(row_heights)
        
        fig = plt.figure(figsize=(total_width / 100, total_height / 100))
        
        # Use GridSpec with custom spacing
        gs = GridSpec(2, 2, width_ratios=col_widths, height_ratios=row_heights, wspace=wspace, hspace=hspace, figure=fig)
        
        axes = []
        for i in range(2):
            for j in range(2):
                ax = fig.add_subplot(gs[i, j])
                ax.imshow(imgs[i * 2 + j])
                ax.axis('off')
                axes.append(ax)
        
        # Add titles
        col_centers = [col_widths[0] / 2, col_widths[0] + col_widths[1] / 2]
        for j in range(2):
            x_pos = col_centers[j] / total_width
            fig.text(x_pos, 0.95, col_titles[j], ha='center', va='bottom', fontsize=12, fontweight='bold')
        
        row_centers = [row_heights[0] / 2, row_heights[0] + row_heights[1] / 2]
        for i in range(2):
            y_pos = row_centers[i] / total_height
            fig.text(0.05, y_pos, row_titles[i], ha='center', va='center', rotation='vertical', fontsize=12, fontweight='bold')

    # Save the figure
    full_dir = os.path.join(output_path,filename)
    plt.savefig(full_dir, dpi=300)
    plt.show()
    plt.close()  # Close to free memory

```

```python


def create_image_grid(images, row_titles, col_titles, filename='fig.png', output_path='./', fig_size=(7, 7), dpi=300):
    """
    Creates a 2x2 grid of images with row and column titles. Assumes all images are the same size.
    - images: List of 4 image paths (strings) or PIL Image objects.
    - row_titles: List of 2 strings for row titles (displayed vertically on the left).
    - col_titles: List of 2 strings for column titles (displayed horizontally on top).
    - output_path: Path to save the output image (default: 'grid_output.png').
    - fig_size: Tuple (width, height) for the figure size in inches (default: (10, 10)). This controls the overall size of the output image.
    - dpi: Resolution for saving the image (default: 600). Higher values improve quality but increase file size.
    """
    if len(images) != 4:
        raise ValueError("Exactly 4 images are required for a 2x2 grid.")
    if len(row_titles) != 2 or len(col_titles) != 2:
        raise ValueError("Exactly 2 row titles and 2 column titles are required.")
    
    # Load images (no resizing)
    imgs = []
    for img in images:
        if isinstance(img, str):
            img = Image.open(img)
        imgs.append(np.array(img))
    
    # Create figure and subplots with no spacing (since images have padding)
    fig, axes = plt.subplots(2, 2, figsize=fig_size, gridspec_kw={'wspace': 0, 'hspace': 0})
    for i in range(2):
        for j in range(2):
            axes[i, j].imshow(imgs[i * 2 + j])
            axes[i, j].axis('off')
    
    # Add column titles (horizontal, centered above each column)
    for j in range(2):
        x_pos = 0.27 + j * 0.44
        fig.text(x_pos, 0.9, col_titles[j], ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # Add row titles (vertical, centered to the left of each row)
    for i in range(2):
        y_pos = 0.72 - i * 0.43
        fig.text(0.07, y_pos, row_titles[i], ha='center', va='center', rotation='vertical', fontsize=12, fontweight='bold')
    
    # Adjust layout
    plt.tight_layout(rect=[0.05, 0.05, 0.95, 0.95])
    
    # Save the figure
    full_dir = os.path.join(output_path,filename)
    plt.savefig(full_dir, bbox_inches='tight' ,dpi=300)
    plt.show()
    plt.close()  # Close to free memory

# Example usage:
# images = ['image1.jpg', 'image2.jpg', 'image3.jpg', 'image4.jpg']
# row_titles = ['Row 1', 'Row 2']
# col_titles = ['Col 1', 'Col 2']
# create_image_grid(images, row_titles, col_titles)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

def create_image_grid(images, row_titles, col_titles, filename='fig.png', output_path='./', fig_size=(7, 10), dpi=300, bottom_title=''):
    """
    Creates a 2x3 grid of images (2 columns, 3 rows) with row and column titles, and an optional bottom title. Assumes all images are the same size.
    - images: List of 6 image paths (strings) or PIL Image objects.
    - row_titles: List of 3 strings for row titles (displayed vertically on the left).
    - col_titles: List of 2 strings for column titles (displayed horizontally on top).
    - output_path: Path to save the output image (default: 'grid_output.png').
    - fig_size: Tuple (width, height) for the figure size in inches (default: (10, 10)). This controls the overall size of the output image.
    - dpi: Resolution for saving the image (default: 600). Higher values improve quality but increase file size.
    - bottom_title: String for a single title displayed at the bottom center (default: ''). If empty, no title is added.
    """
    if len(images) != 6:
        raise ValueError("Exactly 6 images are required for a 2x3 grid.")
    if len(row_titles) != 3 or len(col_titles) != 2:
        raise ValueError("Exactly 3 row titles and 2 column titles are required.")
    
    # Load images (no resizing)
    imgs = []
    for img in images:
        if isinstance(img, str):
            img = Image.open(img)
        imgs.append(np.array(img))
    
    # Create figure and subplots with no spacing (since images have padding)
    fig, axes = plt.subplots(3, 2, figsize=fig_size, gridspec_kw={'wspace': -0.08, 'hspace': -0.08})
    for i in range(3):
        for j in range(2):
            axes[i, j].imshow(imgs[i * 2 + j])
            axes[i, j].axis('off')
    
    # Add column titles (horizontal, centered above each column, closer to grid)
    for j in range(2):
        x_pos = 0.3 + j * 0.4
        fig.text(x_pos, 0.88, col_titles[j], ha='center', va='bottom', fontsize=12, fontweight='bold')
    
    # Add row titles (vertical, centered to the left of each row, closer to grid)
    row_y_positions = [0.77, 0.5, 0.23]  # Evenly spaced for 3 rows
    for i in range(3):
        fig.text(0.1, row_y_positions[i], row_titles[i], ha='center', va='center', rotation='vertical', fontsize=12, fontweight='bold')
    
    # Add bottom title if provided
    if bottom_title:
        fig.text(0.5, 0.1, bottom_title, ha='center', va='top', fontsize=12, fontweight='bold')
    
    # Adjust layout to bring titles closer and leave space for bottom title
    plt.tight_layout(rect=[0.08, 0.08, 0.92, 0.92])
    
    # Save the figure
    full_dir = os.path.join(output_path,filename)
    plt.savefig(full_dir, bbox_inches='tight', dpi=300)
    plt.show()
    plt.close()  # Close to free memory

# Example usage:
# images = ['image1.jpg', 'image2.jpg', 'image3.jpg', 'image4.jpg', 'image5.jpg', 'image6.jpg']
# row_titles = ['Row 1', 'Row 2', 'Row 3']
# col_titles = ['Col 1', 'Col 2']
# create_image_grid(images, row_titles, col_titles, bottom_title='Overall Title')
```

```python


def create_plot_grid(cases, row_titles, col_titles, save_dir, filename='fig.png', output_path='./', fig_size=(10,15), dpi=300, bottom_title=''):
    """
    Creates a 2x3 grid of plots (2 columns, 3 rows) generated from data, with row and column titles, and an optional bottom title.
    Each subplot includes its own colorbar.
    - cases: List of 6 tuples, each (i_case, spin_case, Br, Bth, Bphi) for the data files.
    - row_titles: List of 3 strings for row titles (displayed vertically on the left).
    - col_titles: List of 2 strings for column titles (displayed horizontally on top).
    - save_dir: Directory where the data files are saved.
    - output_path: Path to save the output image (default: './').
    - fig_size: Tuple (width, height) for the figure size in inches (default: (8, 12)).
    - dpi: Resolution for saving the image (default: 300).
    - bottom_title: String for a single title displayed at the bottom center (default: ''). If empty, no title is added.
    """
    if len(cases) != 6:
        raise ValueError("Exactly 6 cases are required for a 2x3 grid.")
    if len(row_titles) != 3 or len(col_titles) != 2:
        raise ValueError("Exactly 3 row titles and 2 column titles are required.")
    
    # Create figure and subplots
    fig, axes = plt.subplots(3, 2, figsize=fig_size, gridspec_kw={'wspace': 0.3, 'hspace': 0.23})
    
    # Reserve space for titles using subplots_adjust (before adding titles)
    plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)  # Reserves space: left 15%, right 15%, top 15%, bottom 15%
    
    for idx, (i_case, spin_case, Br, Bth, Bphi) in enumerate(cases):
        row = idx // 2
        col = idx % 2
        ax = axes[row, col]
        
        # Load r and phi values exported from Mathematica (isco)
        r_phi_data_isco = pd.read_csv("./r_phi_values/r_phi_data_isco_a_%s.csv" % (spin_case), header=None, names=["t", "r", "phi"])
        time_values_isco = r_phi_data_isco["t"].values
        r_values_isco = r_phi_data_isco["r"].values
        phi_values_isco = (r_phi_data_isco["phi"].values) % (2 * np.pi)

        # Load r and phi values exported from Mathematica (plunge)
        r_phi_data_plunge = pd.read_csv("./r_phi_values/r_phi_data_plunge_a_%s.csv" % (spin_case), header=None, names=["t", "r", "phi"])
        time_values_plunge = r_phi_data_plunge["t"].values
        r_values_plunge = r_phi_data_plunge["r"].values
        phi_values_plunge = (r_phi_data_plunge["phi"].values) % (2 * np.pi)

        # Load QU data (using the save_dir parameter, not redefining it)
        filename_plunge = f"i{i_case}_s{spin_case}_B({Br},{Bth},{Bphi})_plungeQU.txt"
        filename_isco = f"i{i_case}_s{spin_case}_B({Br},{Bth},{Bphi})_iscoQU.txt"
        plunge_dir = os.path.join(save_dir, filename_plunge)
        isco_dir = os.path.join(save_dir, filename_isco)
        loadQU_isco = np.loadtxt(isco_dir, unpack=True)
        loadQU_plunge = np.loadtxt(plunge_dir, unpack=True)
        
        Q_plunge, U_plunge = loadQU_plunge
        Q_isco, U_isco = loadQU_isco
        
        # Plotting code adapted for the subplot
        phi_subset_degrees = np.degrees(phi_values_plunge)
        xlim, ylim = auto_limits(Q_plunge, U_plunge, Q_isco, U_isco)
        
        # Create segments for LineCollection
        points = np.array([Q_plunge, U_plunge]).T.reshape(-1, 1, 2)
        segments = np.concatenate([points[:-1], points[1:]], axis=1)
        
        # Create a continuous colormap
        cmap = plt.get_cmap('hsv')
        
        # Create a LineCollection object
        lc = LineCollection(segments, cmap=cmap, linewidth=1)
        lc.set_array(phi_subset_degrees)
        ax.add_collection(lc)
        
        # Plot ISCO
        ax.plot(Q_isco, U_isco, linestyle='-', color="k", linewidth=0.5, label="ISCO", zorder=2)
        
        # Highlight the ϕ=0 point
        x0, y0 = Q_plunge[0], U_plunge[0]
        ax.scatter(x0, y0, s=10, marker="X", color="r", label="ϕ = 0.0" + r"$^{\circ}$", zorder=4)
        
        # Set limits and labels
        ax.set_xlim(xlim)
        ax.set_ylim(ylim)
        ax.set_xlabel("Q", fontsize=10)
        ax.set_ylabel("U", fontsize=10)#, rotation=0)
        ax.grid(alpha=0.2)
        ax.legend(loc='upper right', bbox_to_anchor=(1, 1), frameon=False, fontsize=10)
        
        # # Add legend only to the first subplot
        # if idx == 0:
        #     ax.legend(loc='upper right', bbox_to_anchor=(1, 1), frameon=False, fontsize=10)
        
        # Add colorbar to each subplot (positioned inside the subplot's area, avoiding title spaces)
        # Use the subplot's bbox to position the colorbar within it
        bbox = ax.get_position()
        cbar_ax = fig.add_axes([bbox.x0 + 0.1, bbox.y0 + 0.02, bbox.width * 0.6, 0.007])  # Adjusted width to be proportional, height as in your code
        cbar = plt.colorbar(lc, cax=cbar_ax, orientation="horizontal")
        cbar_ax.text(-0.08, -1.1, "ϕ", transform=cbar_ax.transAxes, ha='right', va='center', fontsize=8)
        cbar.set_ticks(np.arange(0, 361, 90))
        cbar.set_ticklabels([f"{t}°" for t in np.arange(0, 361, 90)], fontsize=8)
    
    # Add column titles (in the reserved top space)
    for j in range(2):
        x_pos = 0.3 + j * 0.4  # Adjust x_pos to fit within the right margin (0.85)
        fig.text(x_pos, 0.86, col_titles[j], ha='center', va='bottom', fontsize=16, fontweight='bold')  # Moved y to 0.92 (within top 15%)
    
    # Add row titles (in the reserved left space)
    row_y_positions = [0.76, 0.51, 0.26]  # Adjusted to fit within the bottom/top margins
    for i in range(3):
        fig.text(0.05, row_y_positions[i], row_titles[i], ha='center', va='center', rotation='vertical', fontsize=16, fontweight='bold')  # Moved x to 0.08 (within left 15%)
    
    # Add bottom title if provided (in the reserved bottom space)
    if bottom_title:
        fig.text(0.5, 0.11, bottom_title, ha='center', va='top', fontsize=12, fontweight='bold')  # Moved y to 0.08 (within bottom 15%)
    
    # No tight_layout here, as subplots_adjust already handles spacing
    
    # Save the figure
    full_path = os.path.join(output_path, filename)
    plt.savefig(full_path, bbox_inches='tight', dpi=dpi)  # Uncommented save, as per your original intent
    plt.show()
    plt.close()


```

```python


def plot_vector_field_subplot(ax, fig, idx, spin_case, r_values_plunge, phi_values_plunge, R, grid_points=16):
    """
    Adapted vector field plot for a subplot.
    """
    # Find range
    x0, y0 = (r_values_plunge * np.cos(phi_values_plunge))[0], (r_values_plunge * np.sin(phi_values_plunge))[0]
    x_range = (-x0 * 1.5, x0 * 1.5)
    y_range = (-x0 * 1.5, x0 * 1.5)
    
    # Create grid
    x = np.linspace(x_range[0], x_range[1], grid_points)
    y = np.linspace(y_range[0], y_range[1], grid_points)
    X, Y = np.meshgrid(x, y)
    
    # Calculate polar coordinates
    r = np.sqrt(X**2 + Y**2)
    theta = np.arctan2(Y, X)
    
    # Define regions
    rH = 1 + np.sqrt(1 - spin_case**2)
    mask_inside = (r < R) & (r > rH)
    mask_outside = r >= R
    mask_excluded = r <= rH
    
    # Vector field functions (assuming isco = R)
    def my_U_inside(X, Y, r, theta):
        omega = (2*spin_case*np.sqrt(R)*(R - r) + R**2*(-2 + r) + spin_case**2*r)/(spin_case**3*r + (-2 + R)*np.sqrt(R)*r**3 + spin_case*(-2*R**2 + r**3) + spin_case**2*np.sqrt(R)*(-2*r + R*(2 + r)))
        iota = -((-1 + R/r)**1.5/(np.sqrt(-1 + (3*R)/2.)*(1 + (2*(1 + (-((spin_case*(spin_case**2 - 2*spin_case*np.sqrt(R) + R**2))/(spin_case + (-2 + R)*np.sqrt(R))) + 2*r)/(spin_case**2 + (-2 + r)*r)))/r)))
        return iota*np.cos(theta) - r*omega*np.sin(theta)
    
    def my_V_inside(X, Y, r, theta):
        omega = (2*spin_case*np.sqrt(R)*(R - r) + R**2*(-2 + r) + spin_case**2*r)/(spin_case**3*r + (-2 + R)*np.sqrt(R)*r**3 + spin_case*(-2*R**2 + r**3) + spin_case**2*np.sqrt(R)*(-2*r + R*(2 + r)))
        iota = -((-1 + R/r)**1.5/(np.sqrt(-1 + (3*R)/2.)*(1 + (2*(1 + (-((spin_case*(spin_case**2 - 2*spin_case*np.sqrt(R) + R**2))/(spin_case + (-2 + R)*np.sqrt(R))) + 2*r)/(spin_case**2 + (-2 + r)*r)))/r)))
        return iota*np.sin(theta) + r*omega*np.cos(theta)
    
    def my_U_outside(X, Y, r, theta):
        omega = 1/(spin_case + r**1.5)
        return -r*omega*np.sin(theta)
    
    def my_V_outside(X, Y, r, theta):
        omega = 1/(spin_case + r**1.5)
        return r*omega*np.cos(theta)
    
    # Calculate vectors
    U_inside = my_U_inside(X, Y, r, theta)
    V_inside = my_V_inside(X, Y, r, theta)
    U_outside = my_U_outside(X, Y, r, theta)
    V_outside = my_V_outside(X, Y, r, theta)
    
    U_combined = np.where(mask_inside, U_inside, np.where(mask_outside, U_outside, np.nan))
    V_combined = np.where(mask_inside, V_inside, np.where(mask_outside, V_outside, np.nan))
    
    # Plot vectors
    quiver = ax.quiver(X, Y, U_combined, V_combined, angles='xy', scale_units='xy', scale=None, pivot='mid',alpha=0.3)
    
    # Add boundaries
    isco_boundary = plt.Circle((0, 0), R, fill=False, color='Black', linestyle='-', linewidth=1, zorder=137)
    ax.plot(np.nan,np.nan,color='Black',linestyle='-', linewidth=1, label=r'$r_{\rm{ISCO}}$')
    rh_boundary = plt.Circle((0, 0), rH, fill=True, color='Black', linestyle=':', linewidth=0.5, zorder=137)
    ax.add_patch(isco_boundary)
    ax.add_patch(rh_boundary)
    
    # Plot trajectory
    points = np.array([r_values_plunge * np.cos(phi_values_plunge), r_values_plunge * np.sin(phi_values_plunge)]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    cmap = plt.get_cmap('hsv')
    phi_subset_degrees = np.degrees(phi_values_plunge)
    lc = LineCollection(segments, cmap=cmap, linewidth=1)
    lc.set_array(phi_subset_degrees)
    ax.add_collection(lc)
    
    # Highlight ϕ=0 point
    ax.scatter(x0, y0, s=10, marker="X", color="r", label="ϕ = 0.0" + r"$^{\circ}$", zorder=137)
    
    # Add colorbar (fixed positioning within subplot)
    bbox = ax.get_position()
    cbar_ax = fig.add_axes([bbox.x0 + 0.1, bbox.y0 + 0.02, bbox.width * 0.6, 0.01])
    cbar = plt.colorbar(lc, cax=cbar_ax, orientation="horizontal")
    cbar_ax.text(-0.08, -1.2, "ϕ", transform=cbar_ax.transAxes, ha='right', va='center', fontsize=8)
    cbar.set_ticks(np.arange(0, 361, 90))
    cbar.set_ticklabels([f"{t}°" for t in np.arange(0, 361, 90)], fontsize=8)
    
    # Configure plot
    ax.plot(0, 0, "xk")
    ax.set_xlabel(r"$X$" + " " + "(M)", fontsize=10)
    ax.set_ylabel(r"$Y$" + " " + "(M)", fontsize=10)
    ax.tick_params(axis='both', which='major', labelsize=10)
    ax.grid(alpha=0.2)
    ax.axis('equal')
    ax.legend(loc='upper right', bbox_to_anchor=(1, 1), frameon=True, fontsize=10, prop={'size': 8})
    # if idx == 0:  # Add legend only to first vector field subplot
    #     ax.legend(loc='upper right', bbox_to_anchor=(1, 1), frameon=True, fontsize=10, prop={'size': 8})

def create_2x2_grid(cases, row_titles, col_titles, save_dir, filename='fig.png', output_path='./', fig_size=(10, 10), dpi=300):
    """
    Creates a 2x2 grid of plots: First column vector fields, second column QU plots.
    - cases: List of 4 tuples, each (i_case, spin_case, Br, Bth, Bphi) for the data files.
    - row_titles: List of 2 strings for row titles (corresponding to the two spin_cases in cases).
    - col_titles: List of 2 strings for column titles.
    - save_dir: Directory where the data files are saved.
    """
    if len(cases) != 4:
        raise ValueError("Exactly 4 cases are required for a 2x2 grid.")
    if len(row_titles) != 2 or len(col_titles) != 2:
        raise ValueError("Exactly 2 row titles and 2 column titles are required.")
    
    fig, axes = plt.subplots(2, 2, figsize=fig_size, gridspec_kw={'wspace': 0.25, 'hspace': 0.22})
    plt.subplots_adjust(left=0.15, right=0.85, top=0.85, bottom=0.15)
    
    for idx, (i_case, spin_case, Br, Bth, Bphi) in enumerate(cases):
        row = idx // 2
        col = idx % 2
        ax = axes[row, col]
        
        # Load common data
        r_phi_data_plunge = pd.read_csv(f"./r_phi_values/r_phi_data_plunge_a_{spin_case}.csv", header=None, names=["t", "r", "phi"])
        r_values_plunge = r_phi_data_plunge["r"].values
        phi_values_plunge = (r_phi_data_plunge["phi"].values) % (2 * np.pi)
        
        if col == 0:  # Vector field column
            plot_vector_field_subplot(ax, fig, idx, spin_case, r_values_plunge, phi_values_plunge, r_values_plunge[0]/0.95)
        else:  # QU plot column
            # Load QU data
            filename_plunge = f"i{i_case}_s{spin_case}_B({Br},{Bth},{Bphi})_plungeQU.txt"
            filename_isco = f"i{i_case}_s{spin_case}_B({Br},{Bth},{Bphi})_iscoQU.txt"
            plunge_dir = os.path.join(save_dir, filename_plunge)
            isco_dir = os.path.join(save_dir, filename_isco)
            loadQU_plunge = np.loadtxt(plunge_dir, unpack=True)
            loadQU_isco = np.loadtxt(isco_dir, unpack=True)
            
            Q_plunge, U_plunge = loadQU_plunge
            Q_isco, U_isco = loadQU_isco
            
            phi_subset_degrees = np.degrees(phi_values_plunge)
            xlim, ylim = auto_limits(Q_plunge, U_plunge, Q_isco, U_isco)
            
            points = np.array([Q_plunge, U_plunge]).T.reshape(-1, 1, 2)
            segments = np.concatenate([points[:-1], points[1:]], axis=1)
            cmap = plt.get_cmap('hsv')
            lc = LineCollection(segments, cmap=cmap, linewidth=1)
            lc.set_array(phi_subset_degrees)
            ax.add_collection(lc)
            
            ax.plot(Q_isco, U_isco, linestyle='-', color="k", linewidth=0.5, label="ISCO", zorder=2)
            x0, y0 = Q_plunge[0], U_plunge[0]
            ax.scatter(x0, y0, s=10, marker="X", color="r", label="ϕ = 0.0" + r"$^{\circ}$", zorder=4)
            
            ax.set_xlim(xlim)
            ax.set_ylim(ylim)
            ax.set_xlabel("Q", fontsize=10)
            ax.set_ylabel("U", fontsize=10)#, rotation=0)
            ax.grid(alpha=0.2)
            
            # Colorbar for QU plot
            bbox = ax.get_position()
            cbar_ax = fig.add_axes([bbox.x0 + 0.1, bbox.y0 + 0.03, bbox.width * 0.6, 0.01])
            cbar = plt.colorbar(lc, cax=cbar_ax, orientation="horizontal")
            cbar_ax.text(-0.08, 0, "ϕ", transform=cbar_ax.transAxes, ha='right', va='center', fontsize=8)
            cbar.set_ticks(np.arange(0, 361, 90))
            cbar.set_ticklabels([f"{t}°" for t in np.arange(0, 361, 90)], fontsize=8)
            ax.legend(loc='upper right', bbox_to_anchor=(1, 1), frameon=False, fontsize=10)
            
            # if idx == 1:  # Legend for first QU plot
            #     ax.legend(loc='upper right', bbox_to_anchor=(1, 1), frameon=False, fontsize=10)
    
    # Add titles
    for j in range(2):
        fig.text(0.3 + j * 0.39, 0.87, col_titles[j], ha='center', va='bottom', fontsize=12, fontweight='bold')
    for i in range(2):
        fig.text(0.08, 0.7 - i * 0.38, row_titles[i], ha='center', va='center', rotation='vertical', fontsize=12, fontweight='bold')
    
    full_path = os.path.join(output_path, filename)
    plt.savefig(full_path, bbox_inches='tight', dpi=dpi)
    plt.show()
    plt.close()

```
