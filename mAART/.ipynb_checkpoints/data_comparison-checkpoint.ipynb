```python
# Find limits for QU loop plots

def auto_limits(x1, y1, x2, y2, padding_factor=0.25, default_lim=(-0.2, 0.2)):
    """
    Automatically calculate xlim and ylim values for two sets of data with divergence handling.
    
    Parameters:
    x1 (array-like): x-coordinate data for first set
    y1 (array-like): y-coordinate data for first set
    x2 (array-like): x-coordinate data for second set
    y2 (array-like): y-coordinate data for second set
    padding_factor (float): percentage of range to add as padding (default: 0.25)
    default_lim (tuple): fallback limits when data is divergent (default: (-0.2, 0.2))
    
    Returns:
    tuple: (xlim, ylim) where each is a (min, max) tuple
    """
    # Convert to numpy arrays and take real parts
    x1 = np.asarray(x1).real
    y1 = np.asarray(y1).real
    x2 = np.asarray(x2).real
    y2 = np.asarray(y2).real
    
    def calculate_lim(data):
        # Handle empty data
        if len(data) == 0:
            return default_lim
        
        # Handle constant values
        if np.all(data == data[0]):
            center = data[0]
            return (center - 1, center + 1)  # Symmetric range around single value
        
        # Handle infinite or NaN values
        if not np.all(np.isfinite(data)):
            return default_lim
            
        # Calculate regular limits with padding
        data_min, data_max = np.nanmin(data), np.nanmax(data)
        data_range = data_max - data_min
        
        # Handle extremely large ranges (divergent cases)
        if data_range > 1e6:  # Adjust threshold as needed
            return default_lim
            
        padding = data_range * padding_factor
        return (data_min - padding, data_max + padding)
    
    # Combine x data from both sets for overall xlim
    xlim = calculate_lim(np.concatenate([x1, x2]))
    
    # Combine y data from both sets for overall ylim
    ylim = calculate_lim(np.concatenate([y1, y2]))
    
    return xlim, ylim
```

```python
fig, ax = plt.subplots(figsize=[6,6])

load_data = np.loadtxt("./experimentalDataQU_t_Q_U.txt",unpack=True)
t_load,Q_load,U_load=load_data
timeInHrs = t_load * t_from_M_to_hours

sc=plt.scatter(Q_load,U_load,c=t_load,s=2,marker="*",cmap="rainbow")
xlim,ylim = auto_limits(Q_load,U_load,Q_load,U_load)

x0, y0 = Q_load[0], U_load[0]
ax.scatter(x0, y0, s=50, marker="X", color="purple", label="t = 0.0", zorder=4)

# Values are in figure fraction coordinates: [left, bottom, width, height]
cbar_ax = fig.add_axes([0.28, 0.19, 0.6, 0.02])  # [x, y, width, height]
cbar = plt.colorbar(sc, cax=cbar_ax, orientation="horizontal")
cbar.set_label("Time [M]")
cbar.set_ticks(np.arange(t_load[0], t_load[-1], 100))
cbar.set_ticklabels([f"{t}" for t in np.arange(t_load[0], t_load[-1], 100)])

# Use twinx() for a vertical colorbar, twiny() for a horizontal one
cax2 = cbar.ax.twiny()

# 2. Set the limits of the new axis
# The limits must correspond to the data range for the secondary labels
cax2.set_xlim(cbar.ax.get_xlim()) # Match the primary axis limits

# 3. Set custom ticks and labels for the second axis
secondary_ticks = [100, 200, 300, 400, 500, 600] 
secondary_labels = ['35.3', '70.6', '105.8', '141.1', '176.4', '211.7']
cax2.set_xticks(secondary_ticks)
cax2.set_xticklabels(secondary_labels)
cax2.set_xlabel('[minutes]') # Add a label for the second axis

# Optional: Adjust the position of the new labels
cax2.yaxis.set_ticks_position('left')
cax2.yaxis.set_label_position('left')

# Plot configuration
ax.plot(0, 0, "xk")  # Ensure it's plotted on the main axes
ax.set_xlim(xlim)
ax.set_ylim(ylim[0]-0.04,ylim[1])
ax.set_xlabel("Q", fontsize=12)
ax.set_ylabel("U", fontsize=12)
ax.grid(alpha=0.2)
ax.legend(loc='upper right', bbox_to_anchor=(1, 1),frameon=False, fontsize=12)
ax.set_title("ALMA dataset 11Apr QU Loop")


filename = f"./QU.png"
plt.savefig(filename,dpi=300)
```
